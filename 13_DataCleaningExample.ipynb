```python
# drop rows with >50% missing
threshold = int(dataf.shape[1]*0.5)
dataf = dataf.dropna(thresh=threshold, axis=0) # axis = 0 = row, and axis = 1 = column
```

```python
dataf.info()
```

```python
dataf.info()
```

```python
# accessing dtypes which are float or int (of a certain type)

for col in dataf.select_dtypes(include=["float64", "int64"]).columns:
    dataf[col].fillna(dataf[col].median(), inplace=True)
```
