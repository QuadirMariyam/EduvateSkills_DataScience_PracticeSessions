# Probability Distribution 
This tells us how the values of a random variable are spread out. <br> Ex, if you flip a coin -> probability of heads = 0.5 Tails = 0.5. <br> That's a distribution

## the types of PD
1. Discrete Distribution - values are countable (coin toss, dice rolls, no. of students, etc.)
2. Continuous Distribution - values can take any number in a range (height, weight, temp, etc.)

It has two classification -> Normal Distribution and Binomial Distribution

# Normal Distribution (Continuous Distribution)


1. also called as Bell Curve
2. Symmetrical around the mean 
3. Most natural phenomena follow it: heights, weights, exam scores
4. Defined by mean and standard deviation

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

```python
# generate 1000 data points from a normal distribution
data = np.random.normal(loc=50, scale=10, size=1000) # mean=50, std=10
```

```python
# Plot Histogram and Curve
plt.hist(data, bins=10, density=True, alpha=0.6, color='skyblue')
plt.title("Normal Distributed Dataset")

# Plot probability density function
xmin, xmax = plt.xlim()
x = np.linspace(xmin, xmax, 100)
p = norm.pdf(x, np.mean(data), np.std(data))
plt.plot(x, p, 'k', linewidth=2)

plt.title("Normal Distribution (Exam Marks Example)")
plt.show()
```

# Binomial Distribution (Discrete Distribution)
models the number successes in n independent trials. Each trial has 2 outcome: either success or failure. Ex, flipping a coin, betting, yes/no) <br>
Parameters:
1. n = number of trials
2. p = probability of success

```python
from scipy.stats import binom

# Example: Tossing a coin 10 times, probability of heads = 0.5

n = 10
p = 0.5
x = np.arange(0, n+1) # possible outcomes (0 to 10 heads)
y = binom.pmf(x, n, p) # probability mass function

plt.bar(x, y, color='lightcoral')
plt.title('Binomial Distribution (Coin Toss Example)')
plt.xlabel('Number of Heads')
plt.ylabel('Probability')
plt.show()
```
