# Data Visualization

Helps in decision-making by visually analysing data

1. Categorical Variables/Column

# Bar Plots
-Show frequency/ count of categories

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# sample dataset

data = pd.DataFrame({
    "Course": ["Python", "Python", "ML", "AI", "AI", "ML", "Python", "AI"],
    "Students": [30, 25, 32, 45, 89, 67, 43, 34]
})
```

```python
# bar plot

sns.countplot(x='Course', data=data)
plt.title("Bar Plot: Students in each Course")
plt.show()
```

# Grouped Bar Plots 
-Compare categories across another categorical variable

```python
data = ({
    "Gender":['M', 'F', 'M', 'F', 'M', 'F', 'M', 'F'],
    "Course": ["Python", "Python", "ML", "AI", "AI", "ML", "Python", "AI"]
})

sns.countplot(x='Course', hue='Gender', data=data)
plt.title("Grouped Bar Plot: Course preference by Gender")
plt.show()
```

2. Continuous Variables

# Histograms
Show the spread (distribution) of a numeric variable

```python
import numpy as np
```

```python
marks = np.random.randint(30, 100, 50)
```

```python
sns.histplot(marks, bins=10, kde=True)
plt.title("Histogram: Distribution of marks")
plt.show()
```

```python
# Scatter plots
# - Check relationship between 2 continuous variables

hours = [2, 3, 5, 7, 9, 10, 12, 15]
scores = [30, 35, 50, 65, 78, 85, 88, 95]

sns.scatterplot(x=hours, y=scores)
plt.title("Scatter Plot: Hours studied vs Exam Score")
plt.show()
```
